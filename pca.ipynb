##In this session we are going to implement a Dimensionality Reduction technique called as PCA which stands for Principal Component Analysis

###PCA is used to reduce the high dimensionality of data. 
###applications:
####Because we can only visualize upto 3 Dimensions. 
####High dimensional data is highly prone to overfitting
####It eases the implementation of ML and DL models

###PCA alogorithm:
####-> Scale all the features to be in the same range
####-> Compute Covariance matrix of the data
####-> Compute Eigen values and Eigen vectors with the help of covariance Matrix
####-> Now, select how many fetures to include from the Eigen vectors
####-> Ex: if you want to reduce the data to K dimensions then eig_vec[: , :k]
####-> perform dot product between the matrix obtained in the previous step and feature vector of the first step
####-> Give the product  to the PCA algorithm
####-> Use the result of PCA in any ML algoriithm

```
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
```

```
#loading the boston house dataset from dklearn
x,y = load_boston(return_X_y=True)
x.shape,y.shape
```

```
x = pd.DataFrame(x)
x
```

```
from sklearn.preprocessing import MinMaxScaler
```

```
# scaling the features to be in a range (0, 1)
scale = MinMaxScaler(feature_range=(0,1))
x = pd.DataFrame(scale.fit_transform(x))
```

```
x
```

```
#Computing covaraince matix
cov_mat = x.cov()
```

```
cov_mat.shape
```

```
#computing Eigen vectors
eig_val, eig_vec = np.linalg.eig(np.array(cov_mat))
```

```
reduced_dim = eig_vec[:,:2]
```

```
x_train = np.dot(x, reduced_dim)
```

```
x_train.shape
```

```
from sklearn.decomposition import PCA
```

```
pca = PCA()
res = pd.DataFrame(pca.fit_transform(x_train))
```
